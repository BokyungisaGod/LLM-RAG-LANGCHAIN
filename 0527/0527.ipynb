```python
import pandas as pd

# CSV 파일 경로 설정
file_path = '/Users/bokyung/Desktop/LLM-RAG-LANGCHAIN/LLM RAG Langchain카톡내용.csv'
df = pd.read_csv(file_path)

# CSV 파일의 열 이름 출력
print(df.columns)
```

```python
# DataFrame의 상위 5개 행을 출력하여 데이터 확인
print(df.head())

# 필요한 데이터(프롬프트와 채팅 내용) 추출
prompts = df['User']
messages = df['Message']
```

```python
# 서술형 문장 생성 함수
def create_narrative(prompts, messages):
    narratives = []
    for prompt, message in zip(prompts, messages):
        narrative = f"{prompt}님이 '{message}'라고 말했어요."
        narratives.append(narrative)
    return narratives

# 서술형 문장 생성
narratives = create_narrative(prompts, messages)

# 서술형 문장 출력
for narrative in narratives:
    print(narrative)
```

```python
# 서술형 문장을 DataFrame으로 변환
narratives_df = pd.DataFrame(narratives, columns=['Narrative'])

# 새로운 CSV 파일로 저장
# narratives_df.to_csv('/Users/bokyung/Desktop/LLM-RAG-LANGCHAIN/kakaotalkprompting.csv', index=False)

# 또는 텍스트 파일로 저장
with open('/Users/bokyung/Desktop/LLM-RAG-LANGCHAIN/kakaotalkprompting.txt', 'w', encoding='utf-8') as f:
    for narrative in narratives:
        f.write(narrative + '\n')
```
